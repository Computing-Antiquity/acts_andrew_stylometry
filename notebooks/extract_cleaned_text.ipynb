```python
replace_dict = {
    "μεθ'": "μετά",
    "μεθ’": "μετά",
    "μεθ´": "μετά",
    "μεθ᾽": "μετά",
    "κατ'": "κατά",
    "κατ’": "κατά",
    "κατ´": "κατά",
    "κατ᾽": "κατά",
    "καθ'": "κατά",
    "καθ’": "κατά",
    "καθ´": "κατά",
    "καθ᾽": "κατά",
    "παρ'": "παρά",
    "παρ’": "παρά",
    "παρ´": "παρά",
    "παρ᾽": "παρά",
    "ὑπ'": "ὑπό",
    "ὑπ’": "ὑπό",
    "ὑπ´": "ὑπό",
    "ὑπ᾽": "ὑπό",
    "ὑπʼ": "ὑπό",
    "ὑφ’": "ὑπό",
    "ὑφ’": "ὑπό",
    "ὑφ᾽": "ὑπό",
    "ὑφ´": "ὑπό",
    "ἀφ’": "ἀπό",
    "ἀφ'": "ἀπό",
    "ἵν'": "ἵνα",
    "ἵν’": "ἵνα",
    "ἵν´": "ἵνα",
    "ἵν᾽": "ἵνα",
    "ἀλλ'": "άλλὰ",
    "ἀλλ’": "άλλὰ",
    "ἀλλ´": "άλλὰ",
    "ἀλλ᾽": "άλλὰ",
    "δι'": "διὰ",
    "δι’": "διὰ",
    "δι´": "διὰ",
    "δι᾽": "διὰ",
    "ἀντ'": "ἀντί",
    "ἀντ’": "ἀντί",
    "ἀντ´": "ἀντί",
    "ἀντ᾽": "ἀντί",
    "ἀνθ'": "ἀντί",
    "ἀνθ’": "ἀντί",
    "ἀνθ´": "ἀντί",
    "ἀνθ᾽": "ἀντί",
    "ἐπ'": "ἐπί",
    "ἐπ’": "ἐπί",
    "ἐπ´": "ἐπί",
    "ἐπ᾽": "ἐπί",
    "ἐφ'": "ἐπί",
    "ἐφ’": "ἐπί",
    "ἐφ´": "ἐπί",
    "ἐφ᾽": "ἐπί"
}

from unicodedata import normalize


def clean_up_text(text: str):
    """Replace the different things in the dictionary.
    """
    clean_text = ""
    for word in normalize("NFC", text.lower()).split():
        if word in replace_dict.keys():
            clean_text += replace_dict[word] + " "
        else:
            clean_text += word + " "
            
    return clean_text

    
```

```python
# Load the texts
from glob import glob
from pathlib import Path
from unicodedata import normalize


loaded_texts = {}

for text in glob("../data/raw_txt/*.txt"):
    loaded_texts[Path(text).stem] = normalize("NFC", clean_up_text(open(text, "r").read().replace("\n", " ")))
```

```python
import spacy


nlp = spacy.load('grc_proiel_lg')


cleaned_text = {}

for text_title, text_content in loaded_texts.items():
    cleaned_text[text_title] = []
    parsed_text = nlp(text_content)
    for parsed_value in parsed_text:
        cleaned_text[text_title].append(
            {
                "raw": parsed_value.text,
                "lemma": parsed_value.lemma_,
                "morphology": parsed_value.morph.to_dict(),
                "pos": parsed_value.pos_,
                "tag": parsed_value.tag_
            }
        )


import json

json.dump(cleaned_text, open("../data/cleaned/cleaned_texts.json", "w"), ensure_ascii=False)
```
